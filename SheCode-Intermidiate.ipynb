DATA CLEANING 

```python
# Select duplicate rows except first occurrence based on all columns
duplicate_rows = data[data.duplicated()]
 
print("Duplicate Rows except first occurrence based on all columns are :")
print(duplicate_rows)
```

```python
data.isnull().sum() #print sum of missing values in each column 
```

```python
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis') #i used the seaborn heatmap to visualize null values in each column
```

```python
data = data.drop(['Age','Cabin'], axis=1) # I dropped the age and cabin column because they contain too many missing values 
```

```python
data = data.drop(['Name' ,'Ticket', 'PassengerId'], axis=1) #i dropped the name and tickets column because they are not usef
```

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked']) #i used the fillna function to fill the missing values in the embarked column 
```

```python
#function to print out features that has categorical variable and how categorical variable it has
for col in data.columns:
    if data[col].dtypes == 'object':
        num_of_unique_cat = len (data[col].unique())
        print("Features '{col_name}' has '{unique_cat}' unique categories". format(col_name=col, unique_cat=num_of_unique_cat))
```

FEATURE ENGINEERING 

```python
#using one hot encoding to deal with the categorical variable 
data = pd.get_dummies(data, columns = ['Sex', 'Embarked' ])
```

MODEL 

```python
#separate the target from the actual data
y = data['Survived']  #target
x = data.drop('Survived', axis = 1) #data or predators 
```

```python
#Standardize the data - Feature Scaling
#tandardized_x = preprocessing.scale(x) 
```

```python
#split the data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
lr = LogisticRegression() # algorithm instantiation
lr.fit(x_train, y_train)
# to check for accuracy score (training)
print('Training score is: ',lr.score(x_train, y_train))
pred = lr.predict(x_test) #make predictions
#to check for accuracy score (testing)
print('Testing score is: ',accuracy_score(y_test, pred))

# to check for f1_score
f1_score(y_test, pred)

# check the classification report
classification_report(y_test, pred)
```

```python
rand = RandomForestClassifier(random_state = 42) # algorithm instantiation
rand.fit(x_train, y_train) # model learning
# evaluating the train data using accuracy score 
print('Training score is: ',rand.score(x_train, y_train))

# make your predictions on the test data
pred = rand.predict(x_test)
# evaluate the test data using accuracy score
print('Testing score is: ',accuracy_score(y_test, pred))

# estimate the f1_score of your predictions
f1_score(y_test, pred)
# classification report of your prediction
classification_report(y_test, pred)
```
